```python
import h5py
data3_pos = h5py.File('fof_subhalo_tab_099.Subhalo.SubhaloPos.hdf5',mode = 'r')
data1_mass = h5py.File('fof_subhalo_tab_099.Subhalo.SubhaloMassType.hdf5',mode = 'r')
data2_sfr = h5py.File('fof_subhalo_tab_099.Subhalo.SubhaloSFR.hdf5',mode='r')
data4_color = h5py.File('fof_subhalo_tab_099.Subhalo.SubhaloStellarPhotometrics.hdf5',mode='r')
head1 =data1_mass['Subhalo']
head2 = data2_sfr['Subhalo']
head3 = data3_pos['Subhalo']
head4 = data4_color['Subhalo']
pos = np.array(head3['SubhaloPos'])
sfr = np.array(head2['SubhaloSFR'])
stellarmass = np.array(head1['SubhaloMassType'])[:,4]
color_g = np.array(head4['SubhaloStellarPhotometrics'])[:,4]
color_r = np.array(head4['SubhaloStellarPhotometrics'])[:,5]
stellarmass = stellarmass.reshape(-1,1)
sfr = sfr.reshape(-1,1)
color_g = color_g.reshape(-1,1)
color_r = color_r.reshape(-1,1)
```

```python
percentiles = np.percentile(df['redshift'].values, [25,75])
```

```python
percentiles
```

```python
np.mean(data_f['redshift'])
```

```python
gal_early = data_f[data_f['redshift']>=percentiles[1]]
gal_late = data_f[data_f['redshift']<=percentiles[0]]
```

```python
xyz_early = gal_early[['x','y','z']].values/1000
xyz_late = gal_late[['x','y','z']].values/1000
xyz_all = data_f[['x','y','z']].values/1000
```

```python
bins = np.linspace(1,10,10)
bincenters = 0.5*(bins[1:]+bins[:-1])
```

```python
import halotools.mock_observables
```

```python
corr_early =  halotools.mock_observables.tpcf(xyz_early, bins, period=205, estimator='Landy-Szalay')
corr_late =   halotools.mock_observables.tpcf(xyz_late, bins, period=205, estimator='Landy-Szalay')
corr_all = halotools.mock_observables.tpcf(xyz_all, bins, period=205, estimator='Landy-Szalay')
```

```python
from astroML.correlation import two_point
```

```python
corr_late = two_point(xyz_late,bins)
corr_early = two_point(xyz_early,bins)
corr_all = two_point(xyz_all,bins)
```

```python
plt.plot(bincenters,corr_late,'b*',label = 'late type galaxies')
plt.plot(bincenters,corr_late,'blue')
plt.plot(bincenters,corr_early,'rx',label = 'early type galaxies')
plt.plot(bincenters,corr_early,'red')
plt.plot(bincenters,corr_all,'go',label = 'all galaxies')
plt.plot(bincenters,corr_all,'green')
plt.title('TNG300-1(galaxies with $10^{10}M_0<M_*<10^{10.2}M_0)$')
plt.yscale('log')
plt.xlabel('r(Mpc/h)');
plt.ylabel('two point correlation function');
plt.legend(loc=0);
#plt.savefig('assembly bias.jpg')
```

```python
sfr_late = gal_late['sfr'].values/gal_late['stellarmass'].values
sfr_early = gal_early['sfr'].values/gal_early['stellarmass'].values
color_late = gal_late['g'].values-gal_late['r'].values
color_early = gal_early['g'].values - gal_early['r'].values
```

```python
plt.scatter(color_early,np.log10(sfr_early*1e10),color = 'red',label = 'early type galaxies',s=0.3)
plt.scatter(color_late,np.log10(sfr_late*1e10),color = 'blue',label = 'late type galaxies',s=0.3)
#plt.yscale('log')
```

```python
plt.scatter(color_late,np.log10(sfr_late*1e10),color = 'blue',label = 'late type galaxies',s=0.8)
plt.scatter(color_early,np.log10(sfr_early*1e10),color = 'red',label = 'early type galaxies',s=0.8)
plt.title('TNG300-1(galaxies with $10^{10}M_0<M_*<10^{10.2}M_0)$')
plt.xlabel('g-r')
plt.ylabel('log(SSFR)')
plt.legend(loc=0)
plt.savefig('ssfr-color.jpg')
```
